```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#importing all the tensorflow libraries
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# reading the data from csv file
data = pd.read_csv('Churn_Modelling.csv')
```

```python
# separating independent variables and target variable

x = data.iloc[:,3:13]
y = data.iloc[:,13]
```

```python
x.info()
```

```python
##Geography and gender are important columns and is a categorical data.
##So we need to convert them to categoricals using categorical encoding.


geography = pd.get_dummies(x['Geography'], drop_first=True)
gender = pd.get_dummies(x['Gender'], drop_first=True)
```

```python
geography
```

```python
gender
```

```python
#Conactenating the categorical column
x = pd.concat([x,geography,gender])
```

```python
x.head()
```

```python
x.drop(columns=['Geography','Gender'],axis=1,inplace=True)
```

```python
x.head()
```

```python
#Split the data in test and train set
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101) 
```

```python
from sklearn.preprocessing import StandardScaler
#feature scaling
sc=StandardScaler()
x_test=sc.fit_transform(x_test)
x_train=sc.fit_transform(x_train)
```

```python
x.head()
```

```python
model = Sequential()

#First hidden layer
model.add(Dense(200,activation='sigmoid'))

#Second hidden layer
model.add(Dense(150,activation='sigmoid'))

#Third hidden layer
model.add(Dense(250,activation='sigmoid'))

#Fourth hidden layer
model.add(Dense(100,activation='sigmoid'))

#Fifth hidden layer
model.add(Dense(50,activation='sigmoid'))

#Output layer
model.add(Dense(2))
```

```python
#compile
model.compile(optimizer='SGD',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#train
model.fit(x_train,y_train)
```

```python
history=model.fit(x_train, y_train, epochs=50, batch_size=32, validation_split = 0.2)
```
